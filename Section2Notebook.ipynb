```python
url = "https://www.basketball-reference.com/leagues/NBA_2020.html"

df = pd.read_html(url, header = 0)

temp_df = df[3]

df_west = temp_df[~temp_df["W"].str.contains("Division")]
```

```python
df_west.columns
```

```python
from pandas.core.tools.numeric import to_numeric
df_west[['W', 'L', 'W/L%', 'GB', 'PS/G', 'PA/G', 'SRS']] = df_west[['W', 'L', 'W/L%', 'GB', 'PS/G', 'PA/G', 'SRS']].apply(to_numeric, errors = "coerce")
```

```python
df_west.dtypes
```

```python
# mean of column W (wins)
df_west["W"].mean()
```

```python
# median of column W (wins)
df_west["W"].median()
```

```python
# std. deviation of column W (wins)
df_west["W"].std()
```

```python
# variance of column W (wins)
df_west["W"].var()
```
